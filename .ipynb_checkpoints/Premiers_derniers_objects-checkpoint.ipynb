```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline  
```

```python
FOLDER = 'plots/'
FOLDER2 = 'product_results/'

dfAA = pd.read_csv('Ordre objets passations finales AA.csv', sep=',', delimiter=None, header=1)
dfAB = pd.read_csv('Ordre objets passations finales AB.csv', sep=',', delimiter=None, header=1)
dfBA = pd.read_csv('Ordre objets passations finales BA.csv', sep=',', delimiter=None, header=1) # not last test
dfBB = pd.read_csv('Ordre objets passations finales BB.csv', sep=',', delimiter=None, header=1)
```

```python
dfBA_mod = dfBA#.drop(['Présentation.11','Essai.11'], axis=1)
dfBA_mod['Test.11'] = dfBA_mod['Test.1']
dfBA_mod
```

```python
df = dfAA.copy()
pres_cols = []
essai_cols = []
test_cols = []
for col in df.columns:
    if 'Présentation' in col:
        pres_cols.append(col)
    if 'Essai' in col:
        essai_cols.append(col)
    if 'Test' in col:
        test_cols.append(col)
print(pres_cols, essai_cols, test_cols)
```

```python
big_df = pd.concat((dfAA,dfAB,dfBA_mod,dfBB),axis=1)
big_df.head(5)
```

```python
word_list = ['ca',
             'ép',
 'co',
 'li',
 'mo',
 'al',
 'lu',
 'ti',
 'br',
 'vo',
 'to',
 'si',
 'ch',
 'cl',
 'la',
 'dé',
 'cu',
 'pi',
 'ta',
 'ba']

def word_count(df,word_list):
    words_occurence = []
    for k,word in enumerate(word_list):
        words_occurence.append(df[word].value_counts())
    return words_occurence

words_occurence_essai = word_count(df=big_df.drop(pres_cols+test_cols,axis=1),word_list=word_list)

pd.DataFrame({'words':word_list,'essai':words_occurence_essai,'test':words_occurence_test})
```

```python
# parameters for the primauté effect
nb_words = 5
nb_firsts = nb_words
nb_lasts  = nb_words
nb_middle = nb_words
words_left = 20 - nb_firsts - nb_lasts  - nb_middle
starting_word_middle = nb_firsts + (words_left - nb_middle)//2

def my_func(df,nb_firsts,nb_lasts):
    how_many_first_and_last_in_essai = []
    how_many_middle_in_essai         = []
    how_many_first_and_last_in_test  = []
    how_many_middle_in_test          = []
    
    how_many_first_in_essai = []
    how_many_last_in_essai  = []
    how_many_first_in_test = []
    how_many_last_in_test  = []

    essai_primaute = np.zeros(20)
    test_primaute  = np.zeros(20)
    
    nb_participants = (len(df.columns))//3
    #print (nb_participants)
    for i in range(nb_participants):#(len(df.columns)):

        ###### pour le nb_premiers, nb_derniers
        pres  = df[df.columns[i*3+0]].dropna()
        essai = df[df.columns[i*3+1]].dropna()
        test  = df[df.columns[i*3+2]].dropna()

        pres_first = pres.iloc[:nb_firsts]
        pres_middle = pres.iloc[starting_word_middle:starting_word_middle+nb_middle]
        pres_last  = pres.iloc[-nb_lasts:]  
        
        #print(pres.iloc[9])

        essai_first = essai.iloc[0:nb_firsts]
        essai_last  =  essai.iloc[-nb_lasts:]

        test_first  =  test.iloc[:nb_firsts]
        test_last   =  test.iloc[-nb_lasts:]
        
        how_many_first_in_essai.append(len(set(list(pres_first)).intersection(essai) ) )
        how_many_last_in_essai.append(len(set(list(pres_last)).intersection(essai) ) )
        
        how_many_first_in_test.append(len(set(list(pres_first)).intersection(test) ) )
        how_many_last_in_test.append(len(set(list(pres_last)).intersection(test) ) )
        
        
        how_many_first_and_last_in_essai.append(len(set(list(pres_first)+list(pres_last)).intersection(essai) ) )
        how_many_first_and_last_in_test.append(len(set(list(pres_first)+list(pres_last)).intersection(test) ) )
        how_many_middle_in_essai.append(2*len(set(list(pres_middle)).intersection(essai) ) )
        how_many_middle_in_test.append(2*len(set(list(pres_middle)).intersection(test) ) )


        #### pour la primauté
        for i in range(20): # 20 is the total number of objects
            if pres[i] in list(essai):
                essai_primaute[i] += 1
            if pres[i] in list(test):
                test_primaute[i] += 1
    print ('firsts in essai', np.mean(how_many_first_in_essai ) ) 
    print ('last in essai', np.mean(how_many_last_in_essai ) ) 
    print ('firsts_and_last in essai', np.mean(how_many_first_and_last_in_essai ) ) 
    print ('middle in essai', np.mean(how_many_middle_in_essai ) ) 

    print ('firsts in test', np.mean(how_many_first_in_test ) ) 
    print ('last in test', np.mean(how_many_last_in_test ) ) 
    print ('firsts + lasts in test', np.mean(how_many_first_and_last_in_test ) ) 
    print ('middle in test', np.mean(how_many_middle_in_test ) ) 
    print ('')
    
    return essai_primaute, test_primaute

prim_essai_same = []
prim_test_same = []
prim_essai_diff = []
prim_test_diff = []

prim_essai = []
prim_test = []


for x in [dfAA, dfBB]:
    tmp_essai, tmp_test = my_func(x,nb_firsts, nb_lasts)
    prim_essai_same.append(tmp_essai)
    prim_test_same.append(tmp_test)
    
for x in [dfBA_mod, dfAB]:
    tmp_essai, tmp_test = my_func(x,nb_firsts, nb_lasts)
    prim_essai_diff.append(tmp_essai)
    prim_test_diff.append(tmp_test)
    
print('laaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa')

print('number of words:', nb_words)
my_names = ['dfAA', 'dfBB', 'dfBA_mod', 'dfAB']
for i,x in enumerate([dfAA, dfBB, dfBA_mod, dfAB]):
    print(my_names[i])
    tmp_essai, tmp_test = my_func(x,nb_firsts, nb_lasts)
    prim_essai.append(tmp_essai)
    prim_test.append(tmp_test)
```

```python

essai_primaute_same = np.mean(prim_essai_same,0)
test_primaute_same = np.mean(prim_test_same,0)

essai_primaute_diff = np.mean(prim_essai_diff,0)
test_primaute_diff = np.mean(prim_test_diff,0)

essai_primaute = np.mean(prim_essai,0)
test_primaute = np.mean(prim_test,0)
```

```python
p1 = plt.figure(1, figsize=(15,9))

# linear
plt.subplot(221)
plt.plot(range(1,21),essai_primaute_same/12*100, 'b', label='essai')
plt.plot(range(1,21),test_primaute_same/12*100, 'r', label='test')
plt.title('Meme contexte')
plt.xlabel('Index du mot lors de la présentation')
plt.ylabel('% de rétention du mot')
plt.legend()
plt.xticks(range(1,21))
plt.xlim((1, 20)) 
plt.grid(True)

# linear
plt.subplot(222)
plt.plot(range(1,21),essai_primaute_diff/12*100, 'b', label='essai')
plt.plot(range(1,21),test_primaute_diff/12*100, 'r', label='test')
plt.title('Contexte différent')
plt.xlabel('Index du mot lors de la présentation')
plt.legend(loc='lower right')
plt.xticks(range(1,21))
plt.xlim((1, 20)) 
plt.grid(True)
p1.savefig('primauté_par_context.png')
```
